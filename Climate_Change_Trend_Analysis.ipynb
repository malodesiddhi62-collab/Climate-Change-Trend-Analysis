## Dataset Description

The dataset contains global temperature data from 1850 to present.

This dataset is used to analyze long-term temperature trends and

```python
import pandas as pd

df = pd.read_csv('annual.csv')

df.head()
```

## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Temperature Trend Visualization

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df['Year'], df['Mean'])
plt.xlabel("Year")
plt.ylabel("Global Temperature")
plt.title("Global Temperature Trend Over Time")
plt.show()
```

## Moving Average Analysis

```python
df['Moving_Avg'] = df['Mean'].rolling(window=10).mean()

plt.figure(figsize=(10,5))
plt.plot(df['Year'], df['Mean'], label='Original')
plt.plot(df['Year'], df['Moving_Avg'], label='Moving Average')
plt.legend()
plt.title("Climate Change Trend Analysis")
plt.show()
```

## Prediction Model

```python
from sklearn.linear_model import LinearRegression

X = df[['Year']]
y = df['Mean']

model = LinearRegression()
model.fit(X, y)

pred = model.predict(X)

plt.figure(figsize=(10,5))
plt.plot(df['Year'], y, label='Actual')
plt.plot(df['Year'], pred, label='Predicted')
plt.legend()
plt.title("Temperature Prediction Trend")
plt.show()
```

## Conclusion

The analysis shows a gradual increase in global temperatures over time,
